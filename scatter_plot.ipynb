```python
# Import Dependencies
import numpy as np
from scipy.stats import linregress, pearsonr
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

```

```python
# Load data from CSV files into pandas DataFrames
tsla_df = pd.read_csv("C:/Boot_camp/Project_1_files/TSLA.csv")
nasdaq_df = pd.read_csv("C:/Boot_camp/Project_1_files/Nasdaq.csv")
nvda_df = pd.read_csv("C:/Boot_camp/Project_1_files/NVDA.csv")
bond_df = pd.read_csv("C:/Boot_camp/Project_1_files/5Yr Bond.csv")
```

```python
# Create scatter plot
plt.scatter(nasdaq_df['Close'], tsla_df['Close'], color='blue')
# Add labels and title
plt.ylabel('TSLA  Price')
plt.xlabel('NASDAQ  Price')
plt.title('TSLA vs NASDAQ Closing Prices')
```

```python
# Create scatter plot
plt.scatter(nasdaq_df['Close'], tsla_df['Close'], color='blue')
# Add labels and title
plt.ylabel('TSLA  Price')
plt.xlabel('NASDAQ  Price')
plt.title('TSLA vs NASDAQ Closing Prices')

# Perform linear regression for TSLA
(slope, intercept, rvalue, pvalue, stderr) = linregress(nasdaq_df['Close'], tsla_df['Close'])
# Plot the linear regression line
regress_value = slope * nasdaq_df['Close'] + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(nasdaq_df['Close'],regress_value,"r-")
plt.annotate(line_eq,(1200,0.5),fontsize=15,color="red")
plt.xlabel('NASDAQ Close Price')
plt.ylabel('TSLA Close Price ')
plt.title('TSLA vs NASDAQ Close Prices with Linear Regression')
#plt.legend()
plt.show()
```

```python
#Calculate correlation between NASDAQ and TESLA Stocks
nasdaq = nasdaq_df['Close']
tesla = tsla_df['Close']
correlation = st.pearsonr(nasdaq,tesla)
print(f"The correlation between NASDAQ and Tesla is {round(correlation[0],2)}")
```

```python
# Create scatter plot
plt.scatter(nasdaq_df['Close'], nvda_df['Close'], color='green')
# Add labels and title
plt.ylabel('NVDA  Price')
plt.xlabel('NASDAQ  Price')
plt.title('NVDA vs NASDAQ Closing Prices')

# Perform linear regression for TSLA
slope, intercept, r_value, _, _ = linregress(nasdaq_df['Close'], nvda_df['Close'])
# Plot the linear regression line
plt.plot(nasdaq_df['Close'], slope * nasdaq_df['Close'] + intercept, color='red', label='Linear Regression')
plt.xlabel('NASDAQ Close Price')
plt.ylabel('NVDA Close Price ')
plt.title('NVDA vs NASDAQ Closing Prices with Linear Regression')
plt.legend()
```

```python
#Calculate correlation between NASDAQ and NVDA Stocks
nasdaq = nasdaq_df['Close']
nvda = nvda_df['Close']
correlation = st.pearsonr(nasdaq,nvda)
print(f"The correlation between NASDAQ and NVDA is {round(correlation[0],2)}")
```
