```python
import pandas as pd 
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
from scipy.stats import linregress, pearsonr
```

```python
five_year_bond = pd.read_csv("/Users/mahermubarak/Desktop/Modules /Project 1/5Yr Bond.csv")
nasdaq = pd.read_csv("/Users/mahermubarak/Desktop/Modules /Project 1/Nasdaq (1).csv")
tesla = pd.read_csv("/Users/mahermubarak/Desktop/Modules /Project 1/TSLA.csv")
nvidia = pd.read_csv("/Users/mahermubarak/Desktop/Modules /Project 1/NVDA (2).csv")
```

```python
nasdaq.head()
```

```python
nasdaq_close = nasdaq.iloc[:,1]
nasdaq_close
```

```python
nasdaq_close_log = np.log(nasdaq_close)
nasdaq_close_log.head(2427)
```

```python
tesla.tail()
```

```python
tesla_close = tesla.iloc[:,1]
tesla_close.tail(2427)
```

```python
tesla_close_log = np.log(tesla_close.tail(2427))
tesla_close_log
```

```python
plt.scatter(nasdaq_close_log.tail(2427), tesla_close_log.tail(2427), s=50, c='blue', alpha=0.5)
plt.xlabel("Nasdaq Closing Price")
plt.ylabel("Tesla Closing Price")
plt.title("Scatter Plot of Nasdaq Log Prices vs. Tesla Closing Log Prices")
plt.grid(True) 

slope, intercept, r_value_, _, _ = linregress(nasdaq_close_log.tail(2427), tesla_close_log.tail(2427))

plt.plot(nasdaq_close_log.tail(2427), slope * nasdaq_close_log.tail(2427) + intercept, color='red', label='Linear Regression')

plt.show()
```

```python
corre = st.pearsonr(nasdaq_close_log.tail(2427), tesla_close_log.tail(2427))
print(f"The Correlation Coefficient between NASDAQ Bond and NVDA is {round(corre[0],2)}")
```
